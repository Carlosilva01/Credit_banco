SITUAÇÃO PROBLEMA

Os dados ficticios pertecem a um banco em Taiwan que oferta serviços de crédito.

* Objetivo: apresentar um panorama consolidado sobre a situação dos clientes no banco

1° PASSO:

Entender os dados do DataFrame

Metadados do DataFrame:

O conjunto de dados credit contém informações sobre pagamentos, fatores demográficos, dados de crédito, histórico de pagamentos e extratos de contas de clientes de cartão de crédito em Taiwan de abril de 2005 a setembro de 2005.

O DataFrame contém 25 variáveis:

  * ID: ID de cada cliente.

  * LIMIT_BAL: Valor do crédito concedido em NT (New Taiwan) dólar (inclui crédito individual e familiar/complementar).

  * SEXO: Gênero (1=masculino, 2=feminino).

  * EDUCAÇÃO: (1 = pós-graduação, 2 = universidade, 3 = ensino médio, 4 = outros, 5 = desconhecido, 6 = desconhecido).

  * CASAMENTO: Estado civil (1=casado, 2=solteiro, 3=outros).

  * IDADE: Idade em anos.

  * PAY_0: Status de pagamento em setembro de 2005 .(-1=pagamento devido, 1=atraso de pagamento por um mês, 2=atraso de pagamento por dois meses, … 8=atraso de pagamento por oito meses, 9=atraso de pagamento por nove meses e acima).

  * PAY_2: Status de reembolso em agosto de 2005 (escala igual à acima).

  * PAY_3: Status de reembolso em julho de 2005 (escala igual à acima).

  * PAY_4: Status de reembolso em junho de 2005 (escala igual à acima).

  * PAY_5: Status de reembolso em maio de 2005 (escala igual à acima).

  * PAY_6: Status de reembolso em abril de 2005 (escala igual à acima).

  * BILL_AMT1: Valor do extrato da conta em setembro de 2005 (dólar NT).

  * BILL_AMT2: Valor do extrato da conta em agosto de 2005 (dólar NT).

  * BILL_AMT3: Valor do extrato da conta em julho de 2005 (dólar NT).

  * BILL_AMT4: Valor do extrato da conta em junho de 2005 (dólar NT).

  * BILL_AMT5: Valor do extrato da conta em maio de 2005 (dólar NT).

  * BILL_AMT6: Valor do extrato da conta em abril de 2005 (dólar NT).

  * PAY_AMT1: Valor do pagamento anterior em setembro de 2005 (dólar NT).

  * PAY_AMT2: Valor do pagamento anterior em agosto de 2005 (dólar NT).

  * PAY_AMT3: Valor do pagamento anterior em julho de 2005 (dólar NT).

  * PAY_AMT4: Valor do pagamento anterior em junho de 2005 (dólar NT).

  * PAY_AMT5: Valor do pagamento anterior em maio de 2005 (dólar NT).

  * PAY_AMT6: Valor do pagamento anterior em abril de 2005 (dólar NT).

  * default.payment.next.month: Pagamento padrão (1=sim, 0=não).

```python
# Renomeando as colunas, para melhor visualização

df_renamed = df.rename(columns = {'LIMIT_BAL': 'valor_credito',
                                'SEX' : 'sexo',
                                'EDUCATION': 'escolaridade',
                                'MARRIAGE' : 'estado_civil',
                                'AGE' : 'idade',
                                'PAY_0' : 'status_pag_set_2005',
                                'PAY_2' : 'status_pag_ago_2005',
                                'PAY_3' : 'status_pag_jul_2005',
                                'PAY_4' : 'status_pag_jun_2005',
                                'PAY_5' : 'status_pag_mai_2005',
                                'PAY_6' : 'status_pag_abr_2005',
                                'BILL_AMT1' : 'extrato_set_2005',
                                'BILL_AMT2' : 'extrato_ago_2005',
                                'BILL_AMT3' : 'extrato_jul_2005',
                                'BILL_AMT4' : 'extrato_jun_2005',
                                'BILL_AMT5' : 'extrato_mai_2005',
                                'BILL_AMT6' : 'extrato_abr_2005',
                                'PAY_AMT1' : 'valor_pag_set_2005',
                                'PAY_AMT2' : 'valor_pag_ago_2005',
                                'PAY_AMT3' : 'valor_pag_jul_2005',
                                'PAY_AMT4' : 'valor_pag_jun_2005',
                                'PAY_AMT5' : 'valor_pag_mai_2005',
                                'PAY_AMT6' : 'valor_pag_abr_2005',


    


})
```

```python
df_renamed
```

```python
#Excluindo a coluna default
#axis = 1 especifica que o que será excluído é uma coluna
#inplace = True faz com que a ação de exclusão seja modificada no DF original (não é obrigatório)

df_renamed.drop(['default payment next month'], axis = 1, inplace = True)
```

```python
# nesta visualização não possui mais a coluna default
df_renamed
```

```python
#Localizando colunas e linhas específicas

df_renamed.loc[0:5, ['idade', 'estado_civil']]
```

```python
# Mostrando apenas os resultados em que a escolaridade seja "2" que corresponde a graduação 
df_renamed[df_renamed['escolaridade'] == 2]
```

```python
# vendo se há números nulos 
# False indiica que não é um número nulo
df_renamed.isnull()
```

```python
# Confirmando se não há valores nulos
# como em todas as linhas o valor é zero, então não há valores nulos

df_renamed.isnull().sum()
```

Como haviam apenas 14 números divergentes em escolaridade, alterei para Nan, e exclui as linhas correspondentes.

```python
df_renamed['escolaridade'].replace(0, np.nan, inplace=True)
```

```python
credit = df_renamed[df_renamed['escolaridade'].notna()]
credit
```

```python
#Excluindo valores duplicados

credit.drop_duplicates(subset = 'ID', keep = 'first', inplace = True)
```

```python
#Neste dataset não há valores duplicados, pois continua com a mesma quantidade de linhas iniciais
credit
```

```python
#Agrupando em média os valores de escolaridade
#É possível verificar que quanto maior o nível de escolaridade, maior é o valor de crédito concedido.
#Escolaridade "4" está como outro, podendo ser mestrado ou doutorado

df_grouped = credit.groupby(['escolaridade']).mean().reset_index()
df_grouped
```

Nível de escolaridade:

 * 1 = Pós-graduação
 * 2 = Graduação
 * 3 = Ensino médio
 * 4 = Outros
 * 5 = Desconhecido
 * 6 = Desconhecido

Analisando o valor de crédito concedido de acordo com a escolaridade. Pode-se notar que quanto maior o nível educacional, maior o limite de crédito concedido.

obs: A escolaridade 4 = Outros, pode ser (Não foi especificado pela empresa) pessoas que tem mestrado ou doutorado.

Observando o gráfico é possível notar que há escolaridade no valor 0, onde não há informação passada pelo banco Credit. Sendo assim, para uma analise de dados correta, será alterado todos os níveis de escolaridade igual a 0 para 5, onde é considerado "Desconhecido"

```python
#Analisando em forma de gráfico

plt.figure(figsize = (10, 5))

plt.xlabel('Escolaridade')
plt.ylabel('Valor credito')

plt.title('Valor de crédito concedido de acordo com a escolaridade')

plt.bar(df_grouped['escolaridade'], df_grouped['valor_credito']);
```

Exportando gráfico sobre o extrato bancário dos últimos 4 meses de cada cliente, de acordo com o nível de escolaridade.

* É possível notar que ao passar dos ultimos 4 meses analisados, não houve uma grande mudança entre os extratos das contas bancárias.

* Houve uma leve alteração de aumento no níveis desconhecido 6 no mês de Julho, se equiparando ao nível de escolaridade 5.

```python
fig, axes = plt.subplots(2, 2, figsize = (17, 10))

sns.set_style('dark') #Usado para deixar o fundo do grárifo escuro

sns.barplot(data = df_grouped, x = 'escolaridade', y = 'extrato_jun_2005', ax = axes[0,0])
sns.barplot(data = df_grouped, x = 'escolaridade', y = 'extrato_jul_2005', ax = axes [0,1])
sns.barplot(data = df_grouped, x = 'escolaridade', y = 'extrato_ago_2005', ax = axes[1,0])
sns.barplot(data = df_grouped, x = 'escolaridade', y = 'extrato_set_2005', ax = axes[1,1]);
```

* Analisando valor pago referente aos últimos meses do ano de 2005.

```python
#usando biblioteca plotly

fig = go.Figure()

fig.add_trace(go.Bar( x = df_grouped['escolaridade'],
                      y = df_grouped['valor_pag_jun_2005'],
                      marker_color = 'Red', #Inserindo cor da barra
                      name = 'Valor pago 06/2005'
))

fig.add_trace(go.Bar(x = df_grouped['escolaridade'],
                     y = df_grouped['valor_pag_jul_2005'],
                     marker_color = 'Blue',
                     name = 'Valor pago 07/2005'
    
))


fig.add_trace(go.Bar(x = df_grouped['escolaridade'],
                     y = df_grouped['valor_pag_ago_2005'],
                     marker_color = 'Green',
                     name = 'Valor pago 08/2005'
    
))


fig.add_trace(go.Bar(x = df_grouped['escolaridade'],
                     y = df_grouped['valor_pag_set_2005'],
                     marker_color = 'yellow',
                     name = 'Valor pago 09/2005'

))

fig.update_layout(xaxis = dict(title = 'Escolaridade',
                               titlefont_size = 15),
                  yaxis = dict(title = ('Valor de pagamento em 2005'),
                               titlefont_size = 15)

)
```
